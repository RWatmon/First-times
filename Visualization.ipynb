```python
#import libraries

import pandas as pd

#importing the visualisation library
from matplotlib import pyplot as plt

import numpy as np

#best practices
plt.style.use("seaborn-v0_8-whitegrid")
plt.rcParams['figure.figsize'] = (10,6)
```

```python
try:
  wine_data =pd.read_csv('/content/winequality-red.csv', sep=';')
  print("Dataset loaded sucessfully!")
  print(f"Original shape of the data is: {wine_data.shape}")
except FileNotFoundError:
  print("File not found. Please upload again, or esnure the path is correct.")
```

```python
duplicate_count = wine_data.duplicated().sum()
print(f"The number of duplicates is: {duplicate_count}")
```

```python
new_wine_data = wine_data.drop_duplicates()
print(f"The new shape of the cleaned data set: {new_wine_data.shape}")
```

```python
#Visualisation

new_wine_data['alcohol'].plot.hist(bins=20, edgecolor='purple', title='Distribution of Alcohol Content')
plt.xlabel('Alcohol Content (%)')
plt.ylabel('Number of Wines Frequency')
plt.show
```

```python
#KDE PLOTTING
new_wine_data['alcohol'].plot.kde(title='Smoothed distribution of Alcohol Content')
plt.xlabel('Alcohol Content (%)')
plt.ylabel('Number of Wines Frequency')
plt.show
```

```python
#BOX PLOTS

new_wine_data[['alcohol', 'pH']].plot.box(title='Spread and outliers of Alcohol and PH')
plt.xlabel('Alcohol Content (%)')
plt.ylabel('Value')
plt.show
```

```python
new_wine_data.plot.hexbin(x='alcohol', y='residual sugar', gridsize=25, title='The Density of Alcohol V Residual Sugar',
                          cmap='Oranges')
```
